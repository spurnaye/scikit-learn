# Simple Linear Regression Using Skicit-learn from "Mastering Machine Learning with Scikit-learn"

Simple Linear Regression: models the relationship between a response variable and single explanatory variable.
Simple linear regression assumes that a linear relationship exists between the response variable and explanatory variable;
it models this relationship with a linear surface called a hyperplane

```python
from sklearn.linear_model import LinearRegression
import Quandl
exxon_data =  Quandl.get("WIKI/XOM")
chevron_data =  Quandl.get("WIKI/CVX")
exxon_data.columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'ExDividend', 'SplitRatio', 'AdjOpen', 'AdjHigh', 'AdjLow', 'AdjClose', 'AdjVolume']
chevron_data.columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'ExDividend', 'SplitRatio', 'AdjOpen', 'AdjHigh', 'AdjLow', 'AdjClose', 'AdjVolume']


x = exxon_data.AdjClose.values
c = chevron_data.AdjClose.values
model = LinearRegression()
c = c[0:11345]

x_arr = []
for element in x:
    x_arr.append([element])
    
c_arr = []
for element in c:
    c_arr.append([element])
    
model.fit(x_arr, c_arr)

print 'Chevron price should be: $%.2f' % model.predict([93.78])[0]



from pandas import DataFrame, Series
xom_cvx_regression = DataFrame()

xom_cvx_regression['xom']  = exxon_data.AdjClose
xom_cvx_regression['cvx'] = chevron_data.AdjClose

xom_cvx_regression.fillna(method='pad')
xom_cvx_regression['cvx_predictions'] = 0

try:
    for idx, row in xom_cvx_regression.iterrows():
        if model.predict([row.xom])[0]:
            xom_cvx_regression.ix[idx, 'cvx_predictions'] = model.predict([row.xom])[0]
        else:
            xom_cvx_regression.ix[idx, 'cvx_predictions'] = 0
except KeyError:
    pass


xom_cvx_regression['cvx_predictions_diff'] = xom_cvx_regression.cvx - xom_cvx_regression.cvx_predictions

xom_cvx_2014_records = xom_cvx_regression.select(lambda dt_index:dt_index.year == 2014)
```

R-squared values range from 0 to 100. An R-squared of 100 means that all movements of a security are completely explained by movements in the index. A high R-squared (between 85 and 100) indicates the fund's performance patterns have been in line with the index. A fund with a low R-squared (70 or less) doesn't act much like the index.
R-squared measures how well the observed values of the response variables are predicted by the model. More
concretely, r-squared is the proportion of the variance in the response variable that is explained by the model.
A higher R-squared value will indicate a more useful beta figure. For example, if a fund has an R-squared value of close to 100 but has a beta below 1, it is most likely offering higher risk-adjusted returns. A low R-squared means you should ignore the beta

```python
from sklearn.linear_model import LinearRegression
import Quandl
import pandas as pd
from pandas import DataFrame, Series

def get_regression(security1, security2):
        print "Fetching Yahoo data for security WIKI/{0}".format(security1)
        first_security = Quandl.get('WIKI/{0}'.format(security1))
        print "Fetching Yahoo data for security WIKI/{0}".format(security2)
        second_security = Quandl.get('WIKI/{0}'.format(security2))
        
        print 'changing columns'
        print first_security.columns
        print second_security.columns
        
        first_security.columns = ['Open_sec1', 'High_sec1', 'Low_sec1', 'Close_sec1', 'Volume_sec1', 'ExDividend_sec1', 'SplitRatio_sec1', 'AdjOpen_sec1', 'AdjHigh_sec1', 'AdjLow_sec1', 'AdjustedClose_sec1', 'AdjVolume_sec1']
        second_security.columns =  ['Open_sec2', 'High_sec2', 'Low_sec2', 'Close_sec2', 'Volume_sec2', 'ExDividend_sec2', 'SplitRatio_sec2', 'AdjOpen_sec2', 'AdjHigh_sec2', 'AdjLow_sec2', 'AdjustedClose_sec2', 'AdjVolume_sec2']
        
        merged_data = pd.merge(first_security,second_security , left_index=True, right_index=True, how='outer')
        merged_data = merged_data.dropna()
        
        combined_2014_records = merged_data.select(lambda dt_index:dt_index.year == 2014)

        first_sec_arr = []
        for element in combined_2014_records.AdjustedClose_sec1:
            first_sec_arr.append([element])
    
        secd_sec_arr = []
        for element in combined_2014_records.AdjustedClose_sec2:
            secd_sec_arr.append([element])

        model = LinearRegression()
        model.fit(first_sec_arr, secd_sec_arr)
        
        print 'R-squared: %.4f' % model.score(first_sec_arr, secd_sec_arr)
        
        print 'Adjusted close variance for {0} for 2014 is {1}'.format(security1, combined_2014_records.AdjustedClose_sec1.var())
        print 'Adjusted close variance for {0} for 2014 is {1}'.format(security2, combined_2014_records.AdjustedClose_sec2.var())
        print 'Covariance of {0} and {1} for 2014 is {2}'.format(security1, security2, combined_2014_records.AdjustedClose_sec1.cov(combined_2014_records.AdjustedClose_sec2))
        print 'Correlation of {0} and {1} for 2014 is {2}'.format(security1, security2, combined_2014_records.AdjustedClose_sec1.corr(combined_2014_records.AdjustedClose_sec2))
        
        return model
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import Quandl
import pandas as pd
from pandas import DataFrame, Series
import numpy as np

def get_data(security):
    print "Fetching Quandl data for security WIKI/{0}".format(security)
    security_data = Quandl.get('WIKI/{0}'.format(security))
    print 'changing columns'
    print security_data.columns
    renamed_columns = []
    for column_name in ['Open', 'High', 'Low', 'Close', 'Volume', 'ExDividend', 'SplitRatio', 'AdjOpen', 'AdjHigh', 'AdjLow', 'AdjClose', 'AdjVolume']:
            renamed_columns.append(column_name + '_'+ security)
    security_data.columns = renamed_columns
    security_data = security_data.dropna()
    return security_data

def get_year_by_records(data, year):
    return data.select(lambda dt_index:dt_index.year == year)

def transform_data(data):
    return data.reshape(data.shape[0], 1)

def get_and_process_data(security_symbol):
    data = get_data(security_symbol)
    data_for_year = get_year_by_records(data, 2014)
    return data_for_year

def find_highest_r_sq():
    xom_data = get_and_process_data('XOM')
    cvx_data = get_and_process_data('CVX')
    merged_data = pd.merge(xom_data,cvx_data , left_index=True, right_index=True, how='outer')
    merged_data = merged_data.dropna()
    transformed_xom = transform_data(merged_data.AdjClose_XOM)
    transformed_cvx = transform_data(merged_data.AdjClose_CVX)
        
    for i in arange(2, 10):
        regressor = LinearRegression()
        regressor.fit(transformed_xom, transformed_cvx)
        quadratic_featurizer = PolynomialFeatures(degree=i)
        xom_quadratic = quadratic_featurizer.fit_transform(transformed_xom)
        regressor_quadratic = LinearRegression()
        regressor_quadratic.fit(xom_quadratic, transformed_cvx)
        print 'Simple linear regression r-squared', regressor.score(transformed_xom, transformed_cvx)
        print 'Quadratic regression r-squared', regressor_quadratic.score(xom_quadratic, transformed_cvx)
        
        
find_highest_r_sq()
```
