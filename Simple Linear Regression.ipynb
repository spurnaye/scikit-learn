# Simple Linear Regression Using Skicit-learn from "Mastering Machine Learning with Scikit-learn"

```python
Simple linear regression assumes that a linear relationship exists between the
response variable and explanatory variable; it models this relationship with a linear
surface called a hyperplane
```

```python
from sklearn.linear_model import LinearRegression
import Quandl
exxon_data =  Quandl.get("WIKI/XOM")
chevron_data =  Quandl.get("WIKI/CVX")
exxon_data.columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'ExDividend', 'SplitRatio', 'AdjOpen', 'AdjHigh', 'AdjLow', 'AdjClose', 'AdjVolume']
chevron_data.columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'ExDividend', 'SplitRatio', 'AdjOpen', 'AdjHigh', 'AdjLow', 'AdjClose', 'AdjVolume']
```

```python
x = exxon_data.AdjClose.values
```

```python
c = chevron_data.AdjClose.values
```

```python
model = LinearRegression()
```

```python
x.size
```

```python
c = c[0:11345]
```

```python
c.size
```

```python
x_arr = []
for element in x:
    x_arr.append([element])
```

```python
c_arr = []
for element in c:
    c_arr.append([element])
```

```python
x_arr
```

```python
c_arr
```

```python
model.fit(x_arr, c_arr)
```

```python
print 'Chevron price should be: $%.2f' % model.predict([93.78])[0]
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.xlabel('Exxon Mobile Price')
plt.ylabel('Cheveon Price')
plt.plot(x_arr, c_arr, 'k.')
plt.axis([0, 120, 0, 120])
plt.grid(True)
plt.show()
```

LinearRegression class is an estimator which predicts value based on previous observations. All estimators implement fit() and predict() methods. fit() method learns parameters of model y = beta x + alpha.

Using training data to learn the values of the parameters for simple linear regression that produce the best fitting model is called ordinary least squares or linear least squares.
